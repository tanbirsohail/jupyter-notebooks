First we import matplotlib and set inline plots in our notebook.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
```

```python
# Now we import seaborn
import seaborn as sb
```

```python
sb.__version__
```

```python
df = pd.read_excel('resources/Pokemon.xls')
```

```python
df.head
```

Now that out dataframe is loaded, we need to draw our plots. I will compare matplotlib's scatterplot with seaborn's lmplot.

```python
# Plotting using good old matplotlib
plt.scatter(x=df.Attack, y=df.Defense, c=df.Stage)
plt.xlim(0,None)
plt.xlabel('Attack')
plt.ylim(0,None)
plt.ylabel('Defence')
plt.show()
```

```python
# Now we try seaborn lmplot()
# We can set fitreg = false to prevent it from plotting the regression line
sb.lmplot(x='Attack', y='Defense', data=df)
plt.xlim(0,None)
plt.xlabel('Attack')
plt.ylim(0,None)
plt.ylabel('Defence')
plt.show()
```

```python
sb.lmplot(x='Attack', y='Defense', data=df, fit_reg=False, hue='Stage')
```

```python
# Matplotlib boxplot by a single column
plt.boxplot(x=df.Attack, data=df)
```

```python
# Seaborn plots box plot for entire datasheet with simple command
plt.figure(figsize=(10,6))
filtered_df = df.drop(['No','Total', 'Stage', 'Legendary'], axis=1)
sb.boxplot(data=filtered_df)
```

```python
# Plotting a swarm plot
# Setting the plot size
plt.figure(figsize=(12,7))
sb.swarmplot(x='Type 1', y='Attack', data=df, hue='Type 1', size=9)
# Setting the legend outside the plot, setting location and anchor of legend
plt.legend(bbox_to_anchor=(1, 1), loc=2)
```

### Now we try some heatmaps

```python
# We try out the normal heatmap and plot the correlations
sb.heatmap(filtered_df.corr())
```

### Histograms
Histograms are drawn beautifully in seaborn, have a look

```python
plt.figure(figsize=(10,7))
sb.distplot(filtered_df.Defense)
```

### Simple bar plots
Lets see the bar plots in seaborn

```python
plt.figure(figsize=(10,7))
sb.countplot(x='Type 1', data=df)
# Rotating the x-labels to make them less clumsy
plt.xticks(rotation=-45)
```

### Making some density/contour plots
We make some density plots

```python
plt.figure(figsize=(10,7))
sb.kdeplot(df.Speed, df.HP)
```
