First it is a good idea to keep the plot diagrams inline with the code cells, not having a popup windown open for each and every plot we draw. For this we use the inline magic command.


```python
import numpy as np
import pandas as pd
import seaborn as sb
from matplotlib import rcParams
```

```python
%matplotlib inline
```

Now we set some parameters for plot size and plot style

```python
rcParams['figure.figsize'] = 5, 4
sb.set_style('whitegrid')
```

```python
x = np.linspace(-6, 6, 100)
y = np.sinc(x)
```

Now we create a *_pandas dataframe_* from the two arrays we defined

```python
df = pd.DataFrame({'x': x, 'sinc':y})
```

We draw a basic line plot using the plot function on pandas DataFrame, which comes packaged with matplotlib.

```python
df.plot.line(x='x', y='sinc')
```

Now we will use Matplotlib's libraries. First we import he *pyplot* library from matplotlib.

```python
import matplotlib.pyplot as plt
```

We find all available styles by running the following command

```python
plt.style.available
```

We will apply certain styles and see how it plays out!

```python
plt.style.use('ggplot')
df.plot.line(x='x', y='sinc')
```

Now lets see other styles as well

```python
plt.style.use('fivethirtyeight')
df.plot.line(x='x', y='sinc')
```

Now the following script created several types of line plots as sublots of a single plot and shows how to add labels as well as legends to a plot.

```python
a = np.linspace(0, 2, 100)

plt.plot(a, a, label='linear')
plt.plot(a, a**2, label='quadratic')
plt.plot(a, a**3, label='cubic')
plt.plot(a, np.power(2,a), label='2^ exponential')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Line Plots")

plt.legend()

plt.show()
```

#### Plotting in subplots
First two digits give the dimension of plot grids. The last digit gives the plot we are accessing in the grid.

```python
plt.figure(1)                # the first figure
plt.subplot(222)             # the first subplot in the first figure
plt.plot([1, 2, 3], 'b--')
plt.title('Second box in grid')
plt.subplot(223)             # the second subplot in the first figure, total there are 2x2 = 4 subplots here
plt.plot([4, 5, 6])
plt.title('Third box in grid')

plt.figure(2)                # a second figure, this is an independent grid, not linked with previous one
plt.plot([4, 5, 6, 7, 8, 9], 'go') # 'go' g stands for green and o stands for dots or o's
plt.title('another figure, having no subplots') # by default, plot will be (111) 1X1 grid, and accessing first subplot division in ovearall plot
```

### Ploting a bar chart

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]
```
