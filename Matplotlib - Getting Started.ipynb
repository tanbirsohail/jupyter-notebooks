First it is a good idea to keep the plot diagrams inline with the code cells, not having a popup windown open for each and every plot we draw. For this we use the inline magic command.


```python
%matplotlib inline
```

```python
import numpy as np
```

```python
import pandas as pd
```

```python
x = np.linspace(-6, 6, 100)
y = np.sinc(x)
```

Now we create a *_pandas dataframe_* from the two arrays we defined

```python
df = pd.DataFrame({'x': x, 'sinc':y})
```

We draw a basic line plot using the plot function on pandas DataFrame, which comes packaged with matplotlib.

```python
df.plot.line(x='x', y='sinc')
```

Now we do full use of standalone Matplotlib library. First we import he *pyplot* library from matplotlib.

```python
import matplotlib.pyplot as plt
```

We find all available styles by running the following command

```python
plt.style.available
```

We will apply certain styles and see how it plays out!

```python
plt.style.use('ggplot')
df.plot.line(x='x', y='sinc')
```

Now lets see other styles as well

```python
plt.style.use('seaborn-poster')
df.plot.line(x='x', y='sinc')
```

Now the following script created several types of line plots and shows how to add labels as well as legends to a plot.

```python
a = np.linspace(0, 2, 100)

plt.plot(a, a, label='linear')
plt.plot(a, a**2, label='quadratic')
plt.plot(a, a**3, label='cubic')
plt.plot(a, np.power(2,a), label='2^ exponential')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Line Plots")

plt.legend()

plt.show()
```
